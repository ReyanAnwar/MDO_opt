## Validation of PyGAD

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import pygad
```

```python
def rastrigin(ga_instance, solution, solution_idx):
    fitness = 20 + (solution[0]**2 - 10*np.cos(2*np.pi*solution[0])) + (solution[1]**2 - 10*np.cos(2*np.pi*solution[1]))
    return -fitness  # maximize fitness → minimize function

fig, ax1 = plt.subplots(1,1,figsize=(8,4))

for i in range(5):
    ga = pygad.GA(
        num_generations=500,
        sol_per_pop=30,
        num_parents_mating=5, # bringing down this value helps
        keep_parents=1,
        parent_selection_type='tournament',
        K_tournament=3,
        num_genes=2,
        mutation_num_genes=1,
        gene_space={'low': -5.12, 'high': 5.12},
        fitness_func=rastrigin,
        mutation_type='random',
        crossover_type="uniform"
    )

    ga.run()

    solution, fitness, _ = ga.best_solution()

    print("Best solution:", solution)
    print("Function value:", -fitness)

    ax1.plot(ga.best_solutions_fitness, label=f'Run {i}')

ax1.grid()
ax1.set_ylabel('Fitness')
ax1.set_xlabel('Generation')

plt.show()
```
